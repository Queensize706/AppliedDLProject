```python
import numpy as np
```

```python
allfile_index = ['001', '002', '005', '012', '016', '019', '023', '031', '035', '057', '059', '064', '075', '078', '081', '084', '091', '094', '096', '101', '110']
len(allfile_index)
```

```python
test_index = ['096', '005', '001']
```

```python
for i in test_index:
    allfile_index.remove(i)

len(allfile_index)
```

```python
slide = []
mask = []

for index in allfile_index:
    data = np.load('index/'+index+'id_im.npy')
    lab = np.load('index/'+index+'id_ma.npy')
    for d in data:
        slide.append(index+'_'+d)
    for l in lab:
        mask.append(index+'_'+l)
```

```python
print(len(slide))
slide[0:5]
```

```python
print(len(mask))
mask[0:5]
```

```python
np.save('slide.npy', slide)
np.save('mask.npy', mask)
```
